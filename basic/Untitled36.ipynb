```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel('sample_data/data aktifitas ATM.xlsx')
```

```python
df1
```

```python
# Disini saya hanya akan memakai data Laporan Kerusakan, CSE, Jenis Layanan dan Bulan
df2 = df1[['Laporan Kerusakan', 'CSE','Jenis Layanan','Bulan']]
# Mengetahui info data
df2.info()
# Mengecek seberapa banyak nilai null yang ada pada data
df2.isnull().sum()
```

```python
# Mengisi nilai null dengan kalimat 'data tidak ada'
df2.fillna('data tidak ada', inplace = True)
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
sns.countplot(ax=ax1,x='Laporan Kerusakan', data=df2)
plt.title('Grafik Laporan Kerusakan Tahun 2020', fontsize = 30, color = 'blue')
print(df2.groupby('Laporan Kerusakan')['Laporan Kerusakan'].count())
```

Dari grafik diatas didapatkan bahwa selama rentang tahun 2020, preventive maintenance adalah laporan kerusakan terbanyak. dan yang paling sedikit adalah non Hardware


```python
print(df2.groupby('CSE')['CSE'].count())
fig, ax1 = plt.subplots(figsize=(20,10))
plt.title('Grafik CSE pada Tahun 2020', fontsize = 30, color = 'blue')
sns.countplot(ax=ax1,x='CSE', data=df2)
```

Dari grafik diatas didapatkan bahwa selama rentang tahun 2020, Donny Indra menjadi yang tertinggi dalam data CSE. dan yang paling sedikit adalah non Hardware

```python
print(df2.groupby('Jenis Layanan')['Jenis Layanan'].count())
fig, ax1 = plt.subplots(figsize=(20,10))
plt.title('Grafik Jenis Layanan', fontsize = 30, color = 'blue')
sns.countplot(ax=ax1,x='Jenis Layanan', data=df2)
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
plt.title('Laporan kerusakan untuk tiap CSE', fontsize = 30, color = 'blue')
sns.countplot(ax=ax1,x='Laporan Kerusakan', hue = df2['CSE'], data=df2)
ax1.axhline(0, color="k", clip_on=False)
```

Dari grafik diatas dapat disimpulkan Donny indra menjadi CSE yang paling banyak melakukan tindakan di semua laporan kerusakan

```python
fig, ax1 = plt.subplots(figsize=(20,10))
plt.title('Grafik Laporan kerusakan untuk tiap Jenis Layanan', fontsize = 30, color = 'blue')
sns.countplot(ax=ax1,x='Laporan Kerusakan', hue = df1['Jenis Layanan'], data=df1)
ax1.axhline(0, color="k", clip_on=False)
```

Dari grafik diatas dapat disimpulkan bahwa Jenis Layanan PM1 memiliki nilai tertinggi pada Laporan preventive maintenance, Jenis Layanan Installation menjadi yang tertinggi pada Laporan Kerusakan Installation dan Jenis layanan Error dan Error&PM menjadi yang tertinggi pada Laporan Kerusakan Hardware

```python
fig, ax1 = plt.subplots(figsize=(20,10))
plt.title('Grafik Laporan Kerusakan tiap bulannya', fontsize = 30, color = 'blue')
sns.countplot(ax=ax1,x='Bulan', hue = df2['Laporan Kerusakan'], data=df2)
ax1.axhline(0, color="k", clip_on=False)
```

Disini dapat dilihat bahwa Preventive Maintenance selalu menjadi Laporan kerusakan tertinggi setiap bulannya dan Hardware selalu menjadi Laporan kerusakan yang tertinggi ke 2 setiap bulannya

```python
fig, ax1 = plt.subplots(figsize=(20,10))
plt.title('Grafik Jenis Layanan tiap bulannya', fontsize = 30, color = 'blue')
sns.countplot(ax=ax1,x='Bulan', hue = df2['Jenis Layanan'], data=df2)
ax1.axhline(0, color="k", clip_on=False)
```

Disini PM1 selalu menjadi Jenis Layanan tertinggi setiap bulannya dan untuk ke 2 tertinggi bervariasi. 
- Pada bulan Juli, September Error menjadi yang ke 2 tertinggi. - Pada bulan Maret, April, Juni, Oktober, Januari dan Februari Jenis Layanan PM2 menjadi yang tertinggi. 
- Pada bulan Desember Installation menjadi Jenis layanan ke 2 tertinggi
- Pada bulan Agustus dan Mei Error&PM menjadi Jenis layanan ke 2 tertinggi secara bersamaan
